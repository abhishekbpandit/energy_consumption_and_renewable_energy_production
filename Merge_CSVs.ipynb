```python
df2 = df2[['Total Energy Sales (MWh)', 'Residential Sales (MWh)', 'Commercial Sales (MWh)', 'Industrial Sales (MWh)', 'Peak Demand (MW)', 'Total Customers', 'Reporting Year']]
df2 = df2.groupby('Reporting Year').agg(sum).reset_index()
df2.head()
```

```python
df2.index = pd.to_datetime(df2['Reporting Year'], format='%Y')

# Drop the original year column
df2 = df2.drop('Reporting Year', axis=1)

# Resample to monthly frequency and forward-fill missing values
df2 = df2.resample('M').ffill()

# reset index
df2 = df2.reset_index()

df2.head()
```

```python
df2['Reporting Year'] = df2['Reporting Year'].apply(lambda x: x + pd.tseries.offsets.MonthBegin(1))
df2.head()
```

```python
# Group by center and sum the new usage column, keep the first FY
df_grouped = df3.groupby(['CENTER', 'FY']).agg({'EC_plant_usage': lambda x: x.sum(skipna=True)}).reset_index()

# Pivot with index FY, columns center and values as the new usage
df_pivoted = df_grouped.pivot(index='FY', columns='CENTER', values='EC_plant_usage')

df_pivoted.head()
```

```python
# Convert the date columns to datetime format
df1['Date'] = pd.to_datetime(df1['Date'])
df2['Reporting Year'] = pd.to_datetime(df2['Reporting Year'], format='%Y-%m-%d')
#df3['Date'] = pd.to_datetime(df3['FY'], format='%Y')
df4['DATE'] = pd.to_datetime(df4['DATE'])
df5['DATE'] = pd.to_datetime(df5['YYYYMM'].astype(str), format='%Y%m')
df6['DATE'] = pd.to_datetime(df6['Month'])
df7['DATE'] = pd.to_datetime(df7['observation_date'], format='%Y-%m-%d')
df8['DATE'] = pd.to_datetime(df8['observation_date'], format='%Y-%m-%d')
df9['DATE'] = pd.to_datetime(df9['Yearmon'], format = '%d-%m-%Y')
df10['DATE'] = pd.to_datetime(df10['DATE'])
df11['DATE'] = pd.to_datetime(df11['DATE'])
df12['DATE'] = pd.to_datetime(df12['Date'])

# Merge the dataframes based on their date columns
merged_df = pd.merge(df1, df2, left_on='Date', right_on='Reporting Year', how='inner')
#merged_df = pd.merge(merged_df, df3, on='Date', how='outer')
merged_df = pd.merge(merged_df, df4, left_on='Date', right_on='DATE', how='inner')
merged_df = pd.merge(merged_df, df5, left_on='Date', right_on='DATE', how='inner')
merged_df = pd.merge(merged_df, df6, left_on='Date', right_on='DATE', how='inner')
merged_df = pd.merge(merged_df, df7, left_on='Date', right_on='DATE', how='inner')
merged_df = pd.merge(merged_df, df8, left_on='Date', right_on='DATE', how='inner')
merged_df = pd.merge(merged_df, df9, left_on='Date', right_on='DATE', how='inner')
merged_df = pd.merge(merged_df, df10, left_on='Date', right_on='DATE', how='inner')
merged_df = pd.merge(merged_df, df11, left_on='Date', right_on='DATE', how='inner')
merged_df = pd.merge(merged_df, df12, left_on='Date', right_on='DATE', how='outer')

merged_df.head()
```
